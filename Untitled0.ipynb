```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Brooklyn_Bridge_Automated_Pedestrian_Counts_Demonstration_Project (1).csv")

df['hour_beginning'] = pd.to_datetime(df['hour_beginning'], errors='coerce')

df_2019 = df[(df['location'] == 'Brooklyn Bridge') & (df['hour_beginning'].dt.year == 2019)].copy()

df_2019['day_of_week'] = df_2019['hour_beginning'].dt.dayofweek
df_weekdays = df_2019[df_2019['day_of_week'] < 5]

weekday_avg = df_weekdays.groupby('day_of_week')['Pedestrians'].mean()

plt.figure(figsize=(8,5))
sns.lineplot(x=['Monday','Tuesday','Wednesday','Thursday','Friday'], y=weekday_avg.values, marker='o')
plt.title('Average Pedestrian Count by Weekday (Brooklyn Bridge, 2019)')
plt.ylabel('Average Pedestrian Count')
plt.xlabel('Day of the Week')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_weather = df_2019.dropna(subset=['weather_summary', 'temperature', 'precipitation'])

weather_group = df_weather.groupby('weather_summary')[['Pedestrians', 'temperature', 'precipitation']].mean()

corr_matrix = weather_group.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Weather vs Pedestrian Counts (2019)')
plt.show()

weather_sorted = weather_group.sort_values('Pedestrians', ascending=False)
print(weather_sorted)
```

```python
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df_2019['time_of_day'] = df_2019['hour_beginning'].dt.hour.apply(categorize_time)

time_activity = (
    df_2019.groupby('time_of_day')['Pedestrians']
    .mean()
    .reindex(['Morning', 'Afternoon', 'Evening', 'Night'])
)

plt.figure(figsize=(7,5))
sns.barplot(x=time_activity.index, y=time_activity.values, palette='coolwarm')
plt.title('Average Pedestrian Activity by Time of Day (Brooklyn Bridge, 2019)')
plt.xlabel('Time of Day')
plt.ylabel('Average Pedestrian Count')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

print(time_activity)
```
